```python
import matplotlib.pyplot as plt
import numpy as np
def quadratic_solution(a, b, c, x):
    return a * x**2 + b * x + c
def read_coefficients_from_file(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()
        a, b, c = map(float, lines)
    return a, b, c

filename = 'task1.csv'

a, b, c = read_coefficients_from_file(filename)

x_values = np.linspace(-10, 10, 100)

y_values = quadratic_solution(a, b, c, x_values)

plt.plot(x_values, y_values, label=f'{a}x^2 + {b}x + {c}')

plt.xlabel('Time')
plt.ylabel('Temperature')
plt.title('Input from file ')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


def quadratic_solution(a, b, c, x):
    return a * x**2 + b * x + c


coefficients_set_1 = (1, -2, 1)
coefficients_set_2 = (5, 1, -2)
coefficients_set_3 = (-1, -3, 2)



x_values = np.linspace(-10, 10, 100)


y_values_set_1 = quadratic_solution(*coefficients_set_1, x_values)
y_values_set_2 = quadratic_solution(*coefficients_set_2, x_values)
y_values_set_3 = quadratic_solution(*coefficients_set_3, x_values)

plt.plot()
plt.plot(x_values, y_values_set_1, label=f'{coefficients_set_1[0]}x^2 + {coefficients_set_1[1]}x + {coefficients_set_1[2]}')
plt.plot(x_values, y_values_set_2, label=f'{coefficients_set_2[0]}x^2 + {coefficients_set_2[1]}x + {coefficients_set_2[2]}')
plt.plot(x_values, y_values_set_3, label=f'{coefficients_set_3[0]}x^2 + {coefficients_set_3[1]}x + {coefficients_set_3[2]}')

plt.xlabel('Time')
plt.ylabel('Temperature')
plt.title('Quadratic Solution with Multiple Sets of Coefficients')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_solution(a, b, c, x):
    y = a * x**2 + b * x + c
    return y

def plot_temperature_over_time(a, b, c, time_range):
    time = np.array(time_range)
    temperature = quadratic_solution(a, b, c, time)

    plt.plot(time, temperature, label=f'a={a}, b={b}, c={c}')

def main():
    print("Weather Modeling using Quadratic Solution")


    num_sets = int(input("Enter the number of sets: "))

    for i in range(num_sets):

        if i == 0:
            a = float(input("Enter the value of a: "))
            b = float(input("Enter the value of b: "))
            c = float(input("Enter the value of c: "))
        else:
            a = 10
            b = 5
            c = 70


        time_range = np.linspace(0, 10, 100)


        plot_temperature_over_time(a, b, c, time_range)

    plt.legend()
    plt.title("Temperature Over Time for Multiple Sets")
    plt.xlabel("Time")
    plt.ylabel("Temperature")
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```
